# Hypothesis


- **Hypothesis 1**: We predict that offenders who are closely related to their victims (e.g., family members, intimate partners) are more likely to use weapons like knives or blunt objects, rather than firearms.

  *Reasoning*: When there is a close relationship between the offender and victim, emotions may play a significant role, leading to weapon choices that are more personal or symbolic. Such incidents may also occur in domestic settings where firearms might be less accessible compared to household items.

- **Hypothesis 2**: We predict that the use of firearms is more common in incidents involving offenders and victims who are strangers compared to incidents involving family members or acquaintances.

  *Reasoning*: Firearms are more commonly used in confrontational or premeditated violence, which tends to occur more often between strangers. The social distance between the offender and the victim may contribute to a detachment that makes firearms more likely.

- **Hypothesis 3**: We expect that incidents involving intimate partners, which are often caused by emotions, are more likely to include weapons of opportunity, such as kitchen knives or blunt objects.

  *Reasoning*: Violent incidents in intimate partner settings may escalate suddenly, with offenders resorting to items readily available in the environment, such as those found in the home.

- **Main Hypothesis**: The choice of weapon in homicide incidents is significantly influenced by the relationship between the offender and the victim. Offenders closely related to their victims (e.g., family members, intimate partners) are more likely to use weapons of opportunity (e.g., knives, blunt objects), while offenders who are strangers to their victims are more likely to use firearms.

  *Reasoning*: In cases where the offender and victim share a close or intimate relationship, the homicide may stem from strong emotional responses (e.g., anger, jealousy, fear). These emotions often escalate in domestic settings where the violence is impulsive. The weapons used in such cases are typically objects readily available in the environment, such as kitchen knives, blunt household items, or tools, because they are easily accessible during a sudden confrontation. In contrast, when the offender and victim are strangers, there is often less emotional involvement. The use of firearms in these cases might be driven by practical considerations such as efficiency, distance, and a lack of personal attachment. Firearms can facilitate premeditated acts of violence or make the offenders feel more controlled in confrontational situations, allowing the offender to maintain distance from the victim. Also, the domestic setting of many intimate partner homicides means that household items (e.g., kitchen knives) are often the most accessible weapons. In contrast, homicides involving strangers may occur in public or outside the home, where firearms are more commonly carried and used. The accessibility of firearms, influenced by legal and cultural factors in Alaska, may also impact this dynamic, but the relationship context is likely a key determinant of weapon choice.

### Perform a Chi-Square Test
The Chi-Square test evaluates whether there is a significant association between weapon type and usage count within the 'Family' relationship.
A significant result would lend statistical support to Hypothesis 1, suggesting that weapon choice is influenced by the offender-victim relationship.

```python
# Chi-Square Test
chi2, p, dof, expected = chi2_contingency(cross_tab)
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)
```

### Conduct a Z-test for proportions
The Z-test compares the proportions of weapon usage (Gun vs. Knife) within the 'Family' relationship.
A statistically significant result can further validate Hypothesis 1 by showing that knives or other weapons of opportunity are used disproportionately compared to guns in such relationships.

```python

# Z-Test for Proportion Comparison
gun_count = weapon_counts['Gun']
knife_count = weapon_counts['Knife']
total_count = gun_count + knife_count

z_stat, p_val = proportions_ztest([gun_count, knife_count], [total_count, total_count])
print("\nZ-Test Results for Proportion Comparison:")
print(f"Z-Statistic: {z_stat}")
print(f"p-value: {p_val}")
```

## Second Analysis  - Data Categorization, Exploratory Analysis, and Predictive Modeling for Behavioral Insights

In this section, we preprocess the dataset by categorizing weapons into meaningful groups and classifying relationships between offenders and victims to facilitate deeper analysis. We then conduct exploratory data analysis to uncover patterns and correlations between relationship types and weapon choices. Finally, we utilize a Random Forest model to predict and assess the relationship between offender-victim dynamics and weapon selection, providing actionable behavioral insights.

First, we want to have a first intuition about whether this is a correlation between the weapon types and relationship between the offender and the victim

We first classify Weapons into three categories:

- **Firearms**.

- **Weapons of Opportunity**: knives, blunt objects.

- **Other**.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Categorize weapons into 'Firearms' and 'Weapons of Opportunity'
def classify_weapon(weapon):
    if pd.isna(weapon):  # Handle missing values
        return "Other"
    weapon = weapon.lower()
    if "gun" in weapon or "firearm" in weapon:
        return "Firearm"
    elif "knife" in weapon or "blunt object" in weapon or "cutting" in weapon:
        return "Weapon of Opportunity"
    else:
        return "Other"

# Apply classification to the dataset
homicide_df['Weapon_Type'] = homicide_df['Weapon'].apply(classify_weapon)
```

We also categorize the Relationship column into two groups:

- **Close Relationships**: Includes family members and intimate partners.

- **Stranger Relationships**.

This categorization is crucial to align with the hypothesis that weapon choice differs based on the closeness of the relationship.

```python
# Categorize relationships into 'Close' and 'Stranger'
def categorize_relationship(relationship):
    if pd.isna(relationship):  # Handle missing values
        return "Other"
    close_relationships = [
        "Mother", "Father", "Wife", "Husband", "Son", "Daughter",
        "Sister", "Brother", "Girlfriend", "Boyfriend", "Ex-husband",
        "Ex-wife", "Other family"
    ]
    if relationship in close_relationships:
        return "Close Relationship"
    elif relationship == "Stranger":
        return "Stranger"
    else:
        return "Other"

# Apply relationship categorization
homicide_df['Relationship_Category'] = homicide_df['Relationship'].apply(categorize_relationship)
```

```python
# Debugging: Check if categories are applied correctly
print("Unique Relationship Categories:", homicide_df['Relationship_Category'].unique())
```

```python
# Group data by Relationship and Weapon_Type
relationship_weapon = homicide_df.groupby(['Relationship_Category', 'Weapon_Type']).size().unstack(fill_value=0)
```

```python
# Normalize to percentages within each relationship group
relationship_weapon_percent = relationship_weapon.div(relationship_weapon.sum(axis=1), axis=0) * 100
```

From the pie chart, we can see there is an obvious discrepancy between firearms and weapons of opportunity in stranger relationships, that firearms are used more frequently, but this isn't true for close relationships.
Although we get some intuitions from the pie chart, we cannot exclude the influences from the confounding variables, for example, the results from the close relationships are opposite from our hypothesis.
So we decide to do a random forest model to predict weapon choice based on relationship and confounders.

We use a Random Forest model to predict weapon choice based on the relationship between the offender and the victim.


```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

```

```python
# Select features and target variable
features = ['Relationship_Category']
target = 'Weapon_Type'
```

```python
# Filter data to include only Close and Stranger relationships
homicide_df = homicide_df[homicide_df['Relationship_Category'].isin(['Close Relationship', 'Stranger'])]
```

```python
# Encode Relationship_Category
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_relationship = encoder.fit_transform(homicide_df[['Relationship_Category']])
encoded_columns = encoder.get_feature_names_out(['Relationship_Category'])
```

```python
# Create feature and target variables
X = pd.DataFrame(encoded_relationship, columns=encoded_columns)
y = homicide_df[target].astype('category').cat.codes
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train a Random Forest model
model = RandomForestClassifier(n_estimators=200, max_depth=10,random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict and evaluate the model
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=homicide_df[target].unique()))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Calculate feature importance
importances = model.feature_importances_
importance_df = pd.DataFrame({'Feature': encoded_columns, 'Importance': importances}).sort_values(by='Importance', ascending=False)
print("\nFeature Importance:\n", importance_df)
```

```python
# Visualize confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=homicide_df[target].unique(), yticklabels=homicide_df[target].unique())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Visualize feature importance
importance_df.plot(kind='bar', x='Feature', y='Importance', legend=False)
plt.title("Feature Importance")
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.show()
```

### Analysis of Random Forest Model Performance

The results of the Random Forest model reveal significant issues with its performance. The classification report shows that the model correctly identifies **43 out of 75** actual "Firearm" cases and **36 out of 40** "Other" cases, but it completely fails to classify the "Weapon of Opportunity" category, with precision, recall, and F1-scores for this class remaining at 0.00. This means that while the model can partially differentiate between "Firearm" and "Other," it misclassifies all "Weapon of Opportunity" cases as either "Firearm" or "Other."

The confusion matrix confirms these findings. Among the "Weapon of Opportunity" cases, none are correctly classified, with **19 misclassified as "Other"** and **5 as "Firearm."** This demonstrates the model’s inability to handle the minority class, likely due to severe class imbalance in the dataset. The underrepresentation of "Weapon of Opportunity" cases biases the model towards the dominant classes, "Firearm" and "Other," resulting in poor generalization across categories.

Additionally, the feature importance analysis reveals that the model relies heavily on the **"Relationship_Category_Stranger"** feature, assigning it most of the predictive power. This over-reliance suggests the model is neglecting other potential factors that influence weapon choice, further limiting its ability to generalize across all categories.

Overall, these results indicate several critical issues: class imbalance, insufficient feature diversity, and the model's inability to differentiate "Weapon of Opportunity" cases. Addressing these problems will require balancing the dataset, incorporating additional features, and potentially exploring alternative modeling approaches to improve classification performance.

We performed correlation analysis to examine relationships between weapon choice, offender-victim relationships, and demographic factors like age and sex. Weak correlations between relationships and weapon choice indicated limited direct influence, suggesting other variables may mediate this effect. To explore potential confounders, we stratified data by victim and offender age groups, as well as sex, and observed notable trends: younger individuals were strongly associated with firearms, and male offenders dominated firearm use, while female offenders used a broader range of weapons. These findings suggest the offender-victim relationship is influenced by, and interacts with, demographic factors, such as age and sex.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Load the dataset
data = homicide_df

# Step 1: Preprocessing - Grouping Categories
# Group similar weapon types into broader categories
data['Weapon_Grouped'] = data['Weapon'].replace({
    'Handgun - pistol, revolver, etc': 'Firearm',
    'Rifle': 'Firearm',
    'Shotgun': 'Firearm',
    'Knife or cutting instrument': 'Sharp Object',
    'Blunt object - hammer, club, etc': 'Blunt Object',
    'Personal weapons, includes beating': 'Personal Weapons',
    'Other firearm': 'Firearm',
    'Other gun': 'Firearm',
    'Firearm, type not stated': 'Firearm',
    'Poison - does not include gas': 'Other',
    'Asphyxiation - includes death by gas': 'Other',
    'Drowning': 'Other',
    'Narcotics or drugs, sleeping pills': 'Other',
    'Other or type unknown': 'Other',
    'Pushed or thrown out window': 'Other',
    'Strangulation - hanging': 'Other'
}).fillna('Other')

# Group similar relationship types into broader categories
data['Relationship_Grouped'] = data['Relationship'].replace({
    'Husband': 'Family',
    'Wife': 'Family',
    'Mother': 'Family',
    'Father': 'Family',
    'Son': 'Family',
    'Daughter': 'Family',
    'Brother': 'Family',
    'Sister': 'Family',
    'Other family': 'Family',
    'Acquaintance': 'Known',
    'Friend': 'Known',
    'Neighbor': 'Known',
    'Boyfriend': 'Intimate Partner',
    'Girlfriend': 'Intimate Partner',
    'Stranger': 'Stranger',
    'Other - known to victim': 'Known'
}).fillna('Other')
```

```python
# Step 2: Encoding Categorical Variables for Analysis
data_encoded = data.copy()
for col in ['Weapon_Grouped', 'Relationship_Grouped', 'VicSex', 'OffSex']:
    data_encoded[col + '_encoded'] = data_encoded[col].astype('category').cat.codes
```

```python
# Step 3: Correlation Analysis
# Correlation matrix for numeric variables
correlation_columns = ['Weapon_Grouped_encoded', 'Relationship_Grouped_encoded', 'VicAge', 'OffAge']
correlation_matrix = data_encoded[correlation_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", xticklabels=correlation_columns, yticklabels=correlation_columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 4: Examining Confounding Variables
# Create age groups for victims and offenders
data['VicAgeGroup'] = pd.cut(data['VicAge'], bins=[0, 18, 30, 50, 100], labels=['0-18', '19-30', '31-50', '51+'])
data['OffAgeGroup'] = pd.cut(data['OffAge'], bins=[0, 18, 30, 50, 100], labels=['0-18', '19-30', '31-50', '51+'])

# Crosstab for Victim Age Group and Weapon Grouped
age_weapon_table = pd.crosstab(data['VicAgeGroup'], data['Weapon_Grouped'])
print("\nVictim Age Group vs. Weapon Choice Table:")
print(age_weapon_table)

# Crosstab for Offender Age Group and Weapon Grouped
off_age_weapon_table = pd.crosstab(data['OffAgeGroup'], data['Weapon_Grouped'])
print("\nOffender Age Group vs. Weapon Choice Table:")
print(off_age_weapon_table)

# Crosstab for Victim Sex and Weapon Choice
vicsex_weapon_table = pd.crosstab(data['VicSex'], data['Weapon_Grouped'])
print("\nVictim Sex vs. Weapon Choice Table:")
print(vicsex_weapon_table)

# Crosstab for Offender Sex and Weapon Choice
offsex_weapon_table = pd.crosstab(data['OffSex'], data['Weapon_Grouped'])
print("\nOffender Sex vs. Weapon Choice Table:")
print(offsex_weapon_table)
```

# Discusison and Conclusion

This study sought to investigate how the offender-victim relationship influences weapon choice in homicide cases, building upon previous research and addressing gaps in understanding the dynamics of violent crime. Our hypothesis proposed that offenders in close relationships with their victims, such as family members or intimate partners, are more likely to use weapons of opportunity (e.g., knives and blunt objects), while strangers are more likely to use firearms. The findings from our analysis support this hypothesis to an extent, but they also highlight the complexity of these patterns, suggesting that additional variables, such as demographic and situational factors, play a significant role.

In the first analysis, the descriptive statistics and chi-square analysis demonstrated a significant association between offender-victim relationship categories and weapon types. This aligns with the hypothesis that weapon choice is influenced by interpersonal dynamics. The Z-test for proportions reinforced these findings, emphasizing disparities in the use of firearms versus weapons of opportunity in familial settings. However, this analysis also pointed to the need for a broader perspective, as cultural norms and accessibility might mediate these trends.

Second analysis extended the analysis to include machine learning and stratified correlation studies. While the random forest model revealed a weak predictive relationship between relationship type and weapon choice, subsequent correlation analyses showed that demographic variables such as age and gender mediate this relationship. Younger offenders and victims were strongly associated with firearm use, and male offenders dominated firearm-related homicides, indicating the interplay of demographic factors with weapon choice. These findings partially align with Dr. Ernesto Lee’s observation of firearm prevalence in homicides while integrating Joseph Gonnella’s findings on the predominance of acquaintance relationships. Our results provide a more nuanced understanding by emphasizing the moderating role of age and sex in offender-victim dynamics.

By categorizing weapon types and relationships, this work extends prior research by introducing a more granular view of weapon use in homicides. Dr. Ernesto Lee’s research primarily highlighted overarching trends, such as the high prevalence of firearms in homicides, whereas Joseph Gonnella focused on the prevalence of relationship types. This study bridges these perspectives, incorporating demographic and situational factors to elucidate how weapon choice is shaped by the intersection of relationship closeness and offender-victim characteristics.

These findings contribute to criminological literature by emphasizing the complex interplay between relationships, demographics, and weapon availability. For example, the high prevalence of firearms in stranger-related homicides suggests not only a practical preference for lethal and distanced methods but also highlights broader societal trends, such as access to firearms and their cultural significance. Conversely, the use of knives and blunt objects in close relationships underscores the role of proximity and impulsivity in familial or intimate violence.

While our hypothesis was supported in several contexts, the presence of confounding factors highlights the need for more robust statistical models to isolate these effects. Future research should incorporate additional contextual variables, such as socioeconomic status, regional firearm regulations, and mental health indicators, to develop a more comprehensive understanding of homicide patterns. Expanding the dataset to include more diverse relationship categories and geographic regions would also improve the generalizability of these findings.

In conclusion, this study confirms that offender-victim relationships influence weapon choice, with close relationships favoring weapons of opportunity and stranger relationships favoring firearms. However, the interplay of confounding factors, such as age and sex, suggests that this relationship is part of a broader, more intricate pattern of violence. These findings not only advance the understanding of homicide dynamics but also provide valuable insights for law enforcement and policymakers in developing targeted prevention strategies and interventions to reduce violent crime.
